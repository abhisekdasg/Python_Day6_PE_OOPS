# 1. Write a python program to create a base class "Shape" with methods to calculate area and perimeter. Then, create derived classes "Circle" and "Rectangle" that inherit from the base class and calculate their respective areas and perimeters. Demonstrate their usage in a program.

        1. Create a class for quizzes and questions.

        3. How would you store and retrieve user progress, including quiz history and scores?


```python
import math

class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

while True:
    print("*" * 25)
    print("\nChoose a shape to calculate its area and perimeter:")
    print("1. Circle")
    print("2. Rectangle")
    print("3. Quit")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        radius = float(input("Enter the radius of the circle: "))
        circle = Circle(radius)
        print("\nCircle Area:", circle.area())
        print("Circle Perimeter:", circle.perimeter())
    elif choice == '2':
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))
        rectangle = Rectangle(length, width)
        print("\nRectangle Area:", rectangle.area())
        print("Rectangle Perimeter:", rectangle.perimeter())
    elif choice == '3':
        break
    else:
        print("Invalid choice. Please enter 1, 2, or 3. From the options")
```

# 3. You are tasked with designing a Python class hierarchy for a simple banking system. The system should be able to handle different types of accounts, such as Savings Accounts and Checking Accounts. Both account types should have common attributes like an account number, account holder's name, and balance. However, Savings Accounts should have an additional attribute for interest rate, while Checking Accounts should have an attribute for overdraft limit.
    
    1. Create a Python class called BankAccount with the following attributes and methods:
        a. Attributes: account_number, account_holder_name, balance
        b. Methods: __init__() (constructor), deposit(), and withdraw()

    2. Create two subclasses, Savings Account and CheckingAccount, that inherit from the BankAccount class.

    3. Add the following attributes and methods to each subclass:
        a. Savings Account:
                i. Additional attribute: interest_rate
                ii. Method: calculate_interest(), which calculates and adds interest to the account based on the interest rate.
        b. Checking Account:
            i. Additional attribute: overdraft_limit
            ii. Method: withdraw(), which allows withdrawing money up to the overdraft limit (if available) without additional fees.
    
    4. Write a program that creates instances of both Savings Account and Checking Account and demonstrates the use of their methods.
    
    5. Implement proper encapsulation by making the attributes private where necessary and providing getter and setter methods as needed.

    6. Handle any potential errors or exceptions that may occur during operations like withdrawals, deposits, or interest calculations. 

    7. Provide comments in your code to explain the purpose of each class, attribute, and method.

    Note: Your code should create instances of the classes, simulate transactions, 
          and showcase the differences between Savings Accounts and Checking Accounts.


```python
# Define a base class for BankAccount
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        """
        Constructor to initialize a bank account.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): Initial balance. Defaults to 0.
        """
        self.__account_number = account_number  # Private attribute for account number
        self.__account_holder_name = account_holder_name  # Private attribute for account holder's name
        self.__balance = balance  # Private attribute for account balance

    def deposit(self, amount):
        """
        Deposit money into the account.

        Args:
            amount (float): The amount to deposit.

        Returns:
            str: A message indicating the deposit status.
        """
        try:
            if amount > 0:
                self.__balance += amount
                return f"Deposited {amount}. New balance: {self.__balance}"
            else:
                raise ValueError("Invalid deposit amount.")
        except ValueError as e:
            return str(e)

    def withdraw(self, amount):
        """
        Withdraw money from the account.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            str: A message indicating the withdrawal status.
        """
        try:
            if amount > 0 and amount <= self.__balance:
                self.__balance -= amount
                return f"Withdrew {amount}. New balance: {self.__balance} Rs."
            else:
                raise ValueError("Invalid withdrawal amount or insufficient funds.")
        except ValueError as e:
            return str(e)

    def get_balance(self):
        """
        Get the account balance.

        Returns:
            float: The account balance.
        """
        return self.__balance

    def set_balance(self, new_balance):
        """
        Set a new balance for the account.

        Args:
            new_balance (float): The new balance to set.

        Returns:
            str: A message indicating the success or failure of setting the new balance.
        """
        try:
            if new_balance >= 0:
                self.__balance = new_balance
                return f"New balance set: {self.__balance} Rs."
            else:
                raise ValueError("Balance cannot be negative.")
        except ValueError as e:
            return str(e)

    def get_account_number(self):
        """
        Get the account number.

        Returns:
            str: The account number.
        """
        return self.__account_number

    def get_account_holder_name(self):
        """
        Get the account holder's name.

        Returns:
            str: The account holder's name.
        """
        return self.__account_holder_name


# Define a subclass for SavingsAccount that inherits from BankAccount
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0, interest_rate=0.0):
        """
        Constructor for SavingsAccount with additional interest_rate attribute.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): Initial balance. Defaults to 0.
            interest_rate (float, optional): Interest rate in percentage. Defaults to 0.0.
        """
        super().__init__(account_number, account_holder_name, balance)
        self.__interest_rate = interest_rate  # Private attribute for interest rate

    def calculate_interest(self):
        """
        Calculate and apply interest to the account based on the interest rate.

        Returns:
            str: A message indicating the success of interest calculation and the new balance.
        """
        try:
            interest_amount = self.get_balance() * (self.__interest_rate / 100)
            self.set_balance(self.get_balance() + interest_amount)
            return f"Interest applied. New balance: {self.get_balance()} Rs."
        except ValueError as e:
            return str(e)


# Define a subclass for CheckingAccount that inherits from BankAccount
class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0, overdraft_limit=0):
        """
        Constructor for CheckingAccount with additional overdraft_limit attribute.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float, optional): Initial balance. Defaults to 0.
            overdraft_limit (float, optional): Overdraft limit. Defaults to 0.
        """
        super().__init__(account_number, account_holder_name, balance)
        self.__overdraft_limit = overdraft_limit  # Private attribute for overdraft limit

    def withdraw(self, amount):
        """
        Withdraw money from the account, allowing overdraft up to the overdraft limit.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            str: A message indicating the withdrawal status.
        """
        try:
            if amount > 0 and (self.get_balance() + self.__overdraft_limit) >= amount:
                self.set_balance(self.get_balance() - amount)
                return f"Withdrew {amount}. New balance: {self.get_balance()} Rs."
            else:
                raise ValueError("Invalid withdrawal amount or overdraft limit exceeded.")
        except ValueError as e:
            return str(e)


# Input and interaction with user
if __name__ == "__main__":
    # User input and account creation
    account_type = input("Enter account type (Savings/Checking): ").strip().lower()
    account_number = input("Enter account number: ")
    account_holder_name = input("Enter account holder's name: ")
    initial_balance = float(input("Enter initial balance: "))

    if account_type == "savings":
        interest_rate = float(input("Enter interest rate (%): "))
        account = SavingsAccount(account_number, account_holder_name, initial_balance, interest_rate)
    elif account_type == "checking":
        overdraft_limit = float(input("Enter overdraft limit: "))
        account = CheckingAccount(account_number, account_holder_name, initial_balance, overdraft_limit)
    else:
        print("Invalid account type. Please enter 'Savings' or 'Checking'.")

    # Display account information and provide options to the user
    print(f"\n{account_type.capitalize()} Account Created:")
    print(f"Account Number: {account.get_account_number()}")
    print(f"Account Holder Name: {account.get_account_holder_name()}")
    print(f"Initial Balance: {account.get_balance()} Rs.")

    while True:
        # Interaction options
        print("\nChoose an option:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Calculate Interest (Savings Account)")
        print("5. Exit")
        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == "1":
            amount = float(input("Enter the deposit amount: "))
            result = account.deposit(amount)
            print(result)
        elif choice == "2":
            amount = float(input("Enter the withdrawal amount: "))
            result = account.withdraw(amount)
            print(result)
        elif choice == "3":
            balance = account.get_balance()
            print(f"Current Balance: {balance} Rs.")
        elif choice == "4" and isinstance(account, SavingsAccount):
            result = account.calculate_interest()
            print(result)
        elif choice == "5":
            print("Exited. Goodbye!")
            break
        else:
            print("Invalid choice. Please select a valid option.")
```

# 4. You are developing an employee management system for a company. Ensure that the system utilizes encapsulation and polymorphism to handle different types of employees, such as full-time and part-time employees.
   
    1. Create a base class called "Employee" with private attributes for name, employee ID, and salary. Implement getter and setter methods for these attributes.
   
    2. Design two subclasses, "FullTimeEmployee" and "PartTimeEmployee," that inherit from "Employee." These subclasses should encapsulate specific properties like hours worked (for part-time employees) and annual salary (for full-time employees).
    
    3. Override the salary calculation method in both subclasses to account for different payment structures.
   
    4. Write a program that demonstrates polymorphism by creating instances of both "Full TimeEmployee" and "PartTimeEmployee." Calculate their salaries and display employee information.


```python
# Define a base class "Employee"
class Employee:
    def __init__(self, name, employee_id, salary):
        self.__name = name  # Private attribute for name
        self.__employee_id = employee_id  # Private attribute for employee ID
        self.__salary = salary  # Private attribute for salary

    # Getter method for name
    def get_name(self):
        return self.__name

    # Setter method for name
    def set_name(self, name):
        self.__name = name

    # Getter method for employee ID
    def get_employee_id(self):
        return self.__employee_id

    # Setter method for employee ID
    def set_employee_id(self, employee_id):
        self.__employee_id = employee_id

    # Getter method for salary
    def get_salary(self):
        return self.__salary

    # Setter method for salary
    def set_salary(self, salary):
        self.__salary = salary

# Define a subclass "FullTimeEmployee" that inherits from "Employee"
class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, annual_salary):
        super().__init__(name, employee_id, annual_salary)  # Inheriting attributes from the base class

    # Override the salary calculation method for full-time employees
    def calculate_salary(self):
        return self.get_salary()  # Annual salary remains the same for full-time employees

# Define a subclass "PartTimeEmployee" that inherits from "Employee"
class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, hourly_rate, hours_worked):
        super().__init__(name, employee_id, None)  # Initialize salary as None, to be calculated later
        self.__hourly_rate = hourly_rate  # Private attribute for hourly rate
        self.__hours_worked = hours_worked  # Private attribute for hours worked

    # Getter method for hourly rate
    def get_hourly_rate(self):
        return self.__hourly_rate

    # Setter method for hourly rate
    def set_hourly_rate(self, hourly_rate):
        self.__hourly_rate = hourly_rate

    # Getter method for hours worked
    def get_hours_worked(self):
        return self.__hours_worked

    # Setter method for hours worked
    def set_hours_worked(self, hours_worked):
        self.__hours_worked = hours_worked

    # Override the salary calculation method for part-time employees
    def calculate_salary(self):
        return self.get_hourly_rate() * self.get_hours_worked()  # Salary = Hourly Rate * Hours Worked

# Function to display employee information
def display_employee_info(employee):
    print("\nEmployee Information:")
    print("Employee ID:", employee.get_employee_id())
    print("Name:", employee.get_name())
    print("Salary:", employee.calculate_salary())  # Using polymorphism to calculate salary

# Input and creation of employee instances using a while loop
if __name__ == "__main__":
    employees = []  # List to store employee instances
    
    while True:
        employee_type = input("\nEnter employee type (FullTime/PartTime) or 'exit' to finish: ").strip().lower()
        
        if employee_type == 'exit':
            break
        
        employee_name = input("Enter employee name: ")
        employee_id = input("Enter employee ID: ")

        if employee_type == "fulltime":
            annual_salary = float(input("Enter annual salary: "))
            employee = FullTimeEmployee(employee_name, employee_id, annual_salary)
        elif employee_type == "parttime":
            hourly_rate = float(input("Enter hourly rate: "))
            hours_worked = float(input("Enter hours worked: "))
            employee = PartTimeEmployee(employee_name, employee_id, hourly_rate, hours_worked)
        else:
            print("Invalid employee type. Please enter 'FullTime' or 'PartTime'.")
            continue  # Skip the rest of the loop if the input is invalid

        employees.append(employee)

    # Display employee information for all employees
    for employee in employees:
        display_employee_info(employee)
```

# 5. Library Management System-Scenario: You are developing a library management system where you need to handle books, patrons, and library transactions.

    1. Create a class hierarchy that includes classes for books (e.g., Book), patrons (e.g., Patron), and transactions (e.g., Transaction). Define attributes and methods for each class.

    2. Implement encapsulation by making relevant attributes private and providing getter and setter methods where necessary.

    3. Use inheritance to represent different types of books (e.g., fiction, non-fiction) as subclasses of the Book class. Ensure that each book type can have specific attributes and methods.

    4. Demonstrate polymorphism by allowing patrons to check out and return books, regardless of the book type. 

    5. Implement a method for tracking overdue books and notifying patrons.

    6. Consider scenarios like book reservations, late fees, and library staff interactions in your design.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

class Patron:
    def __init__(self, name):
        self.name = name
        self.checked_out_books = []

    def check_out_book(self, book):
        if book.is_available:
            book.is_available = False
            self.checked_out_books.append(book)
            return f"{self.name} checked out '{book.title}'."
        else:
            return f"'{book.title}' is not available."

    def return_book(self, book):
        if book in self.checked_out_books:
            book.is_available = True
            self.checked_out_books.remove(book)
            return f"{self.name} returned '{book.title}'."
        else:
            return f"'{book.title}' was not checked out by {self.name}."

# Input the number of books and their details
num_books = int(input("Enter the number of books: "))
library_books = []
for i in range(num_books):
    title = input(f"Enter title for book {i + 1}: ")
    author = input(f"Enter author for book {i + 1}: ")
    book = Book(title, author)
    library_books.append(book)

# Input the number of patrons
num_patrons = int(input("Enter the number of patrons: "))
patrons = []
for i in range(num_patrons):
    patron_name = input(f"Enter the name of patron {i + 1}: ")
    patron = Patron(patron_name)
    patrons.append(patron)

# Perform book transactions
while True:
    print("\nSelect a patron for a transaction:")
    for i, patron in enumerate(patrons):
        print(f"{i + 1}. {patron.name}")

    patron_choice = input("Enter the patron number (or 'exit' to finish): ")
    if patron_choice == 'exit':
        break

    try:
        patron_index = int(patron_choice) - 1
        if 0 <= patron_index < num_patrons:
            # Check if any books are available
            available_books = [i for i, book in enumerate(library_books) if book.is_available]
            if not available_books:
                print("All books are checked out.")
                continue  # Go back to the beginning of the loop

            print("Books available in the library:")
            for i, book_index in enumerate(available_books):
                book = library_books[book_index]
                print(f"{i + 1}. {book.title} by {book.author} - Available: {book.is_available}")

            book_choice = input(f"{patrons[patron_index].name}, select a book to check out (enter book number): ")
            book_index = int(book_choice) - 1

            if 0 <= book_index < len(available_books):
                result = patrons[patron_index].check_out_book(library_books[available_books[book_index]])
                print(result)
            else:
                print("Invalid book number.")
        else:
            print("Invalid patron number.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")
```

# 6.Online Shopping Cart
    Scenario: You are tasked with designing a class hierarchy for an online shopping cart system. The system should handle products, shopping carts, and orders. Consider various OOP principles while designing this system.
    
    1. Create a class hierarchy that includes classes for products (e.g., Product), shopping carts (e.g., ShoppingCart), and orders (e.g., Order). Define attributes and methods for each class.
    
    2. Implement encapsulation by making relevant attributes private and providing getter and setter methods where necessary.
    
    3. Use inheritance to represent different types of products (e.g., electronics, clothing) as subclasses of the Product class. Ensure that each product type can have specific attributes and methods.
    
    4. Demonstrate polymorphism by allowing various product types to be added to a shopping cart and calculate the total cost of items in the cart.
    
    5. Implement a method for placing an order, which transfers items from the shopping cart to an order.
    
    Consider scenarios like out-of-stock products, discounts, and shipping costs in your design.


```python
# Define a global dictionary to store product stock quantities
product_stock = {}

class Product:
    def __init__(self, product_id, name, price, stock_quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock_quantity = stock_quantity  # Initialize stock quantity

        # Store the stock quantity in the global dictionary
        product_stock[self.product_id] = self.stock_quantity

    def get_product_id(self):
        return self.product_id

    def get_name(self):
        return self.name

    def get_price(self):
        return self.price

    def get_stock_quantity(self):
        return self.stock_quantity

    def set_stock_quantity(self, quantity):
        self.stock_quantity = quantity
        # Update the global stock quantity
        product_stock[self.product_id] = self.stock_quantity

    def is_available(self):
        return self.stock_quantity > 0

    def __str__(self):
        return f"{self.name} (ID: {self.product_id}) - {self.price} Rs. (Stock: {self.stock_quantity})"

class ElectronicProduct(Product):
    def __init__(self, product_id, name, price, brand, stock_quantity):
        super().__init__(product_id, name, price, stock_quantity)
        self.brand = brand

    def get_brand(self):
        return self.brand

    def __str__(self):
        return super().__str__() + f" - Brand: {self.brand}"

class ClothingProduct(Product):
    def __init__(self, product_id, name, price, size, stock_quantity):
        super().__init__(product_id, name, price, stock_quantity)
        self.size = size

    def get_size(self):
        return self.size

    def __str__(self):
        return super().__str__() + f" - Size: {self.size}"

class ShoppingCart:
    def __init__(self):
        self.items = {}  # Dictionary to store items (product_id: quantity)

    def add_item(self, product, quantity):
        if product.get_product_id() in self.items:
            self.items[product.get_product_id()] += quantity
        else:
            self.items[product.get_product_id()] = quantity

    def remove_item(self, product, quantity):
        if product.get_product_id() in self.items:
            self.items[product.get_product_id()] -= quantity
            if self.items[product.get_product_id()] <= 0:
                del self.items[product.get_product_id()]

    def get_total_cost(self):
        total_cost = 0
        for product_id, quantity in self.items.items():
            product = ProductCatalog.get_product_by_id(product_id)
            if product:
                total_cost += product.get_price() * quantity
        return total_cost

class ProductCatalog:
    catalog = {}  # Dictionary to store products (product_id: Product)

    @classmethod
    def add_product(cls, product):
        cls.catalog[product.get_product_id()] = product

    @classmethod
    def get_product_by_id(cls, product_id):
        return cls.catalog.get(product_id)

class Order:
    def __init__(self, order_id, customer_name, shipping_address, items):
        self.order_id = order_id
        self.customer_name = customer_name
        self.shipping_address = shipping_address
        self.items = items  # List of (product, quantity) tuples

    def calculate_total_cost(self):
        total_cost = 0
        for product, quantity in self.items:
            total_cost += product.get_price() * quantity
        return total_cost

    def place_order(self):
        for product, quantity in self.items:
            if product.is_available():
                product.set_stock_quantity(product.get_stock_quantity() - quantity)
            else:
                print(f"Product {product.get_name()} is out of stock.")

# Function to create a new product based on user input
def create_product():
    product_id = int(input("Enter product ID: "))
    name = input("Enter product name: ")
    price = float(input("Enter product price: "))
    stock_quantity = int(input("Enter product stock quantity: "))
    category = input("Enter product category (Electronic/Clothing): ")

    if category.lower() == "electronic":
        brand = input("Enter brand: ")
        product = ElectronicProduct(product_id, name, price, brand, stock_quantity)
    elif category.lower() == "clothing":
        size = input("Enter size: ")
        product = ClothingProduct(product_id, name, price, size, stock_quantity)
    else:
        print("Invalid category. Product not added.")
        return

    ProductCatalog.add_product(product)
    print(f"Product {product.get_name()} added to the catalog.")

# Function to interactively add products to the shopping cart
def add_to_cart():
    product_id = int(input("Enter the product ID to add to the cart: "))
    quantity = int(input("Enter the quantity: "))
    product = ProductCatalog.get_product_by_id(product_id)
    if product:
        available_stock = product_stock.get(product_id, 0)
        if available_stock >= quantity:
            cart.add_item(product, quantity)
            print(f"{quantity} {product.get_name()}(s) added to the cart.")
        else:
            print(f"Only {available_stock} {product.get_name()}(s) available.")
    else:
        print("Product not found in the catalog.")

# Function to place an order based on the items in the shopping cart
def place_order(cart):
    customer_name = input("Enter your name: ")
    shipping_address = input("Enter your shipping address: ")

    if not cart.items:
        print("Shopping cart is empty. Cannot place an order.")
        return

    order_items = []  # List to store (product, quantity) tuples
    for product_id, quantity in cart.items.items():
        product = ProductCatalog.get_product_by_id(product_id)
        if product and product.is_available():
            available_stock = product_stock.get(product_id, 0)
            if available_stock >= quantity:
                order_items.append((product, quantity))
                # Update the global stock quantity
                product_stock[product_id] -= quantity
            else:
                print(f"Product {product.get_name()} has insufficient stock and will not be included in the order.")
        elif product:
            print(f"Product {product.get_name()} is out of stock.")
        else:
            print(f"Product with ID {product_id} is not found in the catalog and will not be included in the order.")

    if not order_items:
        print("No valid products in the shopping cart. Cannot place an order.")
        return

    order = Order(1, customer_name, shipping_address, order_items)
    order.place_order()
    print(f"Order placed successfully!")

# Function to display all products in the catalog
def view_products():
    print("\nProducts in the Catalog:")
    for product_id, product in ProductCatalog.catalog.items():
        print(product)

# Main program loop
cart = ShoppingCart()  # Initialize the shopping cart
while True:
    print("\nOptions:")
    print("1. Create a new product")
    print("2. Add a product to the shopping cart")
    print("3. View products in the catalog")
    print("4. View shopping cart and calculate total cost")
    print("5. Place an order")
    print("6. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        create_product()
    elif choice == "2":
        add_to_cart()
    elif choice == "3":
        view_products()  # Option to view products
    elif choice == "4":
        total_cost = cart.get_total_cost()
        print(f"Total cost in the cart: {total_cost} Rs.")
    elif choice == "5":
        place_order(cart)  # Pass the cart as an argument
    elif choice == "6":
        break
    else:
        print("Invalid choice. Please try again.")
```
